```python
def preprocessing(dataframe):
    responses = []
    failed_responses = []

    for _, row in dataframe.iterrows():
        prompt = row["question"]
        task_id = row["problem_id"]
        solution = row["solutions"]

        try:
            generated_soln = model.generate_content(prompt)

            if generated_soln.candidates[0].finish_reason == 1:
                print(f"{task_id} done")
                responses.append(
                {
                    "task_id": task_id,
                    "prompt": prompt,
                    "solution": solution,
                    "generated_output": generated_soln.text,
                }
            )
            elif generated_soln.candidates[0].finish_reason == 4:
                print(f"{task_id} generating copyrighted material")
                failed_responses.append(
                {
                    "task_id": task_id,
                    "prompt" : prompt,
                    "solution" : solution,
                    "failed_reason": "Reciting Copyrighted material"
                }
                )
        except Exception as e:
            print(f"Error generating responses for prompt {task_id}\n{e}")
            if "429 Resource has been exhausted" in str(e):
                print("Rate Limit Exceeded. Trying again after cooldown.")
                time.sleep(45)
                try:
                    # Retry the request
                    generated_soln = model.generate_content(prompt)
                    if generated_soln.candidates[0].finish_reason == 1:
                        print(f"{task_id} done (after retry)")
                        responses.append(
                            {
                                "task_id": task_id,
                                "prompt": prompt,
                                "solution": solution,
                                "generated_output": generated_soln.candidates[0].content.parts[0].text,
                            }
                        )
                    else:
                        print(f"{task_id} failed after retry")
                        failed_responses.append(
                            {
                                "task_id": task_id,
                                "prompt": prompt,
                                "solution": solution,
                                "failed_reason": "Failed after retry",
                            }
                        )
                except Exception as retry_error:
                    if "429 Resource has been exhausted" in str(retry_error):
                        print("Daily quota exceeded. Terminating function.")
                        break  # Exit the function if RPD quota is exceeded
                    else:
                        print(f"Unexpected error during retry for {task_id}\n{retry_error}")
                        failed_responses.append({
                            "task_id": task_id,
                            "prompt": prompt,
                            "solution": solution,
                            "failed_reason": e
                        })
            else:
                print(f"Unexpected error for {task_id}\n{e}")
                failed_responses.append({
                    "task_id": task_id,
                    "prompt": prompt,
                    "solution": solution,
                    "failed_reason": e
                })

    return responses, failed_responses 
            
            
```

```python
import numpy as np
import pandas as pd

def split_dataframe(dataframe, batch_size):
    """
    Splits a DataFrame into smaller DataFrames of the specified batch size.

    Args:
        dataframe (pd.DataFrame): The DataFrame to split.
        batch_size (int): The number of rows per batch.

    Returns:
        List[pd.DataFrame]: A list of DataFrame batches.
    """
    num_batches = int(np.ceil(len(dataframe) / batch_size))
    return [dataframe.iloc[i * batch_size: (i + 1) * batch_size] for i in range(num_batches)]
```

```python
remaining_indices = {i for i in range(0,len(df)) if i not in generated_response_indices and i not in failed_response_indices}
len(remaining_indices)
```

```python
# Assuming `remaining_indices` is a list or set of problem_id values
remaining_df = df[df['problem_id'].isin(remaining_indices)]
```

```python
responses = []
failed_responses = []
```

```python
batch_size = 500
batches = split_dataframe(remaining_df, batch_size)
```

```python
for batch in batches:
    batch_responses, batch_failed = preprocessing(batch)  # Updated preprocessing to return both
    responses.extend(batch_responses)  # Add successful responses to main list
    failed_responses.extend(batch_failed)  # Add failed tasks to failed list
```

```python
loaded_responses.extend(responses)
loaded_failed_responses.extend(failed_responses)
```

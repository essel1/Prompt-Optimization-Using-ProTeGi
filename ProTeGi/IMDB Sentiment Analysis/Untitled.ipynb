```python
from sklearn.metrics import accuracy_score
```

```python
from cohere import ClassifyExample

examples = [
    # Positive examples
    ClassifyExample(
        text="This movie was absolutely amazing! The acting and story were top-notch.",
        label="Positive"
    ),
    ClassifyExample(
        text="An extraordinary experience. I loved every second of it.",
        label="Positive"
    ),
    ClassifyExample(
        text="The cinematography and direction were outstanding. Highly recommend!",
        label="Positive"
    ),
    ClassifyExample(
        text="A masterpiece! Easily one of the best films I have seen this year.",
        label="Positive"
    ),
    ClassifyExample(
        text="Heartwarming and beautifully crafted. A must-watch.",
        label="Positive"
    ),
    
    # Negative examples
    ClassifyExample(
        text="The movie was incredibly boring and a complete waste of time.",
        label="Negative"
    ),
    ClassifyExample(
        text="Terrible acting and a confusing plot. I wouldn't recommend this to anyone.",
        label="Negative"
    ),
    ClassifyExample(
        text="Disappointing from start to finish. It lacked depth and emotion.",
        label="Negative"
    ),
    ClassifyExample(
        text="The visuals were subpar, and the story was completely predictable.",
        label="Negative"
    ),
    ClassifyExample(
        text="An unmemorable experience. I struggled to stay awake.",
        label="Negative"
    ),
]
```

```python
label_mapping = { "positive": 1, "negative": 0}
df['sentiment'] = df['sentiment'].map(label_mapping)
```

```python
df = df.iloc[0:20000]
```

```python
def get_predictions(batch_reviews):
    response = co.classify(
        model="embed-english-v3.0",
        inputs= batch_reviews,
        examples = examples
    )
    return [1 if prediction.prediction == 'Positive' else 0 for prediction in response.classifications]
```

```python
batch_size = 95
predictions = []
```

```python
for i in range(0, len(df), batch_size):
    batch_reviews = df['review'].iloc[i:i+batch_size].tolist()
    batch_predictions = get_predictions(batch_reviews)
    predictions.extend(batch_predictions)
```

```python
accuracy = accuracy_score(df['sentiment'], predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

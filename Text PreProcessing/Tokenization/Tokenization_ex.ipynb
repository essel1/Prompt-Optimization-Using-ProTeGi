```python
corpus = open("./corpus.txt").read()
print(corpus)
```

```python
word_str = corpus.split()
```

```python
print(word_str)
```

```python
import spacy
import re
```

```python
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
```

```python
nlp = English()
text_str = nlp(corpus)
words = [ token.text for token in text_str]
print(words)
print(len(words))
```

```python
## White space tokenizer
def white_space_tokenizer(nlp):
    return Tokenizer(nlp.vocab, token_match = re.compile(r'\S+').match)

nlp.tokenizer = white_space_tokenizer(nlp)
text_str_1 = nlp(corpus)
words = [ token.text for token in text_str_1 ]
print(words)
```

```python
print(len(words))
```

```python
def hypenate_tokenizer(nlp):
    return Tokenizer(nlp.vocab, token_match = re.compile(r'\b\W+(?:-\W+)*').match)

nlp.tokenizer = hypenate_tokenizer(nlp)
text_str_2 = nlp(corpus)
words = [ token.text for token in text_str_2 ]
print(words)
```
